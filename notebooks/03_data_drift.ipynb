```python
# ============================================================
# Analyse du Data Drift avec Evidently
# ============================================================
import pandas as pd
import numpy as np
from evidently import Dataset, DataDefinition
from evidently.presets import DataDriftPreset
from evidently import Report
```

```python
# ============================================================
# Chargement des données
# ============================================================

df_train = pd.read_csv('../data/application_train_prepared.csv')
df_test = pd.read_csv('../data/application_test.csv')

print(f"Train shape : {df_train.shape}")
print(f"Test shape  : {df_test.shape}")
print(f"Colonnes en commun : {len(set(df_train.columns) & set(df_test.columns))}")
```

```python
# ============================================================
# Préparation des données pour Evidently
# ============================================================

cols_to_exclude = ['TARGET', 'SK_ID_CURR']
common_cols = list(set(df_train.columns) & set(df_test.columns) - set(cols_to_exclude))
print(f"Colonnes utilisées pour l'analyse : {len(common_cols)}")

train_sample = df_train[common_cols].sample(5000, random_state=42)
test_sample = df_test[common_cols].sample(5000, random_state=42)

print(f"Shape train sample : {train_sample.shape}")
print(f"Shape test sample  : {test_sample.shape}")
```

```python
# ============================================================
# Génération du rapport Data Drift
# ============================================================

data_definition = DataDefinition()

train_dataset = Dataset.from_pandas(train_sample, data_definition=data_definition)
test_dataset = Dataset.from_pandas(test_sample, data_definition=data_definition)

report = Report([DataDriftPreset()])
my_run = report.run(reference_data=train_dataset, current_data=test_dataset)
my_run.save_html('../models/data_drift_report.html')
```
