```python
# ============================================================
# Analyse du Data Drift avec Evidently
# ============================================================
import pandas as pd
import numpy as np
from evidently import Dataset, DataDefinition
from evidently.presets import DataDriftPreset
from evidently import Report
```

```python
df_train = pd.read_csv('../data/application_train_prepared.csv')

# Simulation : référence = données d'entraînement, current = données de production
df_reference = df_train.sample(frac=0.5, random_state=42)
df_current = df_train.drop(df_reference.index)

print(f"Reference shape : {df_reference.shape}")
print(f"Current shape   : {df_current.shape}")
```

```python
cols_to_exclude = ['TARGET', 'SK_ID_CURR']
feature_cols = [c for c in df_train.columns if c not in cols_to_exclude]
print(f"Colonnes analysées : {len(feature_cols)}")

reference_sample = df_reference[feature_cols].sample(5000, random_state=42)
current_sample = df_current[feature_cols].sample(5000, random_state=42)

print(f"Shape reference sample : {reference_sample.shape}")
print(f"Shape current sample   : {current_sample.shape}")
```

```python
data_definition = DataDefinition()

reference_dataset = Dataset.from_pandas(reference_sample, data_definition=data_definition)
current_dataset = Dataset.from_pandas(current_sample, data_definition=data_definition)

report = Report([DataDriftPreset()])
my_run = report.run(reference_data=reference_dataset, current_data=current_dataset)
my_run.save_html('../models/data_drift_report.html')
print("Fichier sauvegardé : models/data_drift_report.html")
```
